```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

# Read the features dataframe from CSV
#average_features_df = pd.read_csv('/content/drive/MyDrive/data/average_features_df.csv')
features_df = pd.read_csv('/content/drive/MyDrive/data/features_df.csv')
playlists_df = pd.read_csv('/content/drive/MyDrive/data/playlist_df.csv')
tracks_df = pd.read_csv('/content/drive/MyDrive/data/tracks_df.csv')
ratings_df = pd.read_csv('/content/drive/MyDrive/data/ratings_df.csv')
artist_info_df = pd.read_csv('/content/drive/MyDrive/data/artist_info_df.csv')
release_year_df = pd.read_csv('/content/drive/MyDrive/data/processed_release_dates.csv')
```

```python
artist_info_df.drop(['external_urls', 'followers', 'href', 'id', 'uri', 'name'], axis=1, inplace=True)
artist_info_df.drop(['images'], axis=1, inplace=True)
artist_info_df.drop(['type'], axis=1, inplace=True)
artist_info_df.rename(columns={'popularity': 'artist_popularity'}, inplace=True)
artist_info_df.head()
# Has artist popularity + genres
```

```python
tracks_df = tracks_df.iloc[:, 1:]
tracks_df = tracks_df.merge(artist_info_df, on='artist_uri', how='left')
release_year_df['album_uri'] = release_year_df['album_uri'].str.split(':').str[-1]
tracks_df = tracks_df.merge(release_year_df, on='album_uri', how='left')
tracks_df.head()
# has info of unique tracks
```

```python
tracks_df = tracks_df.merge(features_df, on='track_id', how='left')
tracks_df.head()
```

```python
tracks_df.drop(columns=["Unnamed: 0"], inplace = True)
tracks_df.isna().sum()
```

```python
tracks_df.dropna(inplace=True)
tracks_df.head()
```

```python
tracks_df.isna().sum()
```

```python
ratings_df = ratings_df.iloc[:, 1:]
ratings_df.head()
# Has playlist to track mapping
```
